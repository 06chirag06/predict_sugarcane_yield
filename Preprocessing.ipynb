```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas.plotting import scatter_matrix

import seaborn as sns
```

```python
production='Sugarcane_final.xlsx'
data = pd.read_excel(production)
data=data.sort_values(by=['Year','Indian State'])
data
```

```python
print("Highest allowed",data['Temperature'].mean() + 3*data['Temperature'].std())
print("Lowest allowed",data['Temperature'].mean() - 3*data['Temperature'].std())
```

```python
new_df = data[(data['Temperature'] < 34.69) & (data['Temperature'] > 14.69)]
data=new_df
data
```

```python
new_df = data[(data['Production'] < 102116.12) & (data['Production'] > 500.0)]
data=new_df
```

```python
print("Highest allowed",data['Rainfall'].mean() + 3*data['Rainfall'].std())
print("Lowest allowed",data['Rainfall'].mean() - 3*data['Rainfall'].std())
```

```python
new_df = data[(data['Rainfall'] < 5966.74) & (data['Rainfall'] > 0.0)]
data=new_df
data
```

```python
data.to_excel('final.xlsx',index=False)
data
```
